```
import pandas as pd
```

```
# 2019년 세대관련연간자료 데이터를 불러온다.
path = '/content/drive/My Drive/Datas/brunch/전출입/세대관련연간자료_2019.csv'
df = pd.read_csv(path, encoding='cp949')
```

```
# 전입행정_시도코드와 전입행정_시군구코드를 합쳐서 5자리의 in_cd 코드 컬럼을 생성한다. 
df['in_cd'] = df['전입행정_시도'].astype(str) + df['전입행정_시군구'].astype(str)

# 전출행정_시도코드와 전출행정_시군구코드를 합쳐서 5자리의 out_cd 코드 컬럼을 생성한다. 
df['out_cd'] = df['전출행정_시도'].astype(str) + df['전출행정_시군구'].astype(str)

df.tail(3)
```

```
# 전출지(in_cd)별 전입지(out_cd)의 이동_총인구의 합계를 구한다.
df_grp = df.groupby(['out_cd', 'in_cd'])['이동_총인구'].sum().reset_index()
df_grp
```

```
# 전출지와 전입지가 같은 행의 인덱스 리스트를 생성한다.
df_grp_idx = df_grp[df_grp.out_cd == df_grp.in_cd].index

# df_grp_idx 리스트를 이용하여 전출지와 전입지가 같은 행을 삭제하다.
df_grp_result = df_grp.drop(index=df_grp_idx).reset_index(drop=True)
df_grp_result
```

```
# 결과를 파일로 저장한다.
df_grp_result.to_csv('/content/drive/My Drive/Datas/brunch/전출입/시군구이동집계.csv')
```
